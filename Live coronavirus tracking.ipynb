```python
import pandas as pd
df = pd.read_json('https://www.mohfw.gov.in/data/datanew.json')
df
```

```python
df = df.iloc[:-1,1:]
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
total_states = np.arange(len(df['state_name']))
```

```python
total_states
```

```python
df['active']
```

```python
max(df['positive'])
```

### Total Positive Cases based on States

```python
plt.figure(figsize=(10,15))
plt.barh(total_states,df['positive'],align='center',alpha=0.5,color=(1,0,0))
plt.yticks(total_states,df['state_name'])
plt.xlim(1,max(df['positive'])+100)
plt.xlabel('Positive Number of Cases')
plt.title('Coronavirus Positive Cases')
plt.show()
```

### Active Number of Cases based on states

```python
plt.figure(figsize=(10,15))
plt.barh(total_states,df['new_active'],align='center',color=(1,0,0))
plt.yticks(total_states,df['state_name'])
plt.xlabel('Active Number of Cases')
plt.title('Coronavirus Active Cases')
plt.xlim(1,max(df['new_active'])+10)
plt.show()
```

```python
plt.figure(figsize=(10,15))
plt.barh(total_states,df['new_death'],color=(0,1,1))
plt.xlim(1,max(df['new_death'])+10)
plt.xlabel("Death Number of Cases")
plt.yticks(total_states,df['state_name'])
plt.show('COronavirus Death cases')
plt.show()
```
